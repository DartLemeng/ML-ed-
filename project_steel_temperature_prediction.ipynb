```python
## функция для переименования столбцов

def rename(df, col_old, col_new):
    if len(col_old) == len(col_new):
        for _ in range(len(col_old)):
            df = df.rename(columns={col_old[_]: col_new[_]})
        return df
    else:
        print('кол-во названий старых и новых столбцов не равно')
```

```python
## функция для переименования столбцов, меняющая буквы верхнего на буквы нижнего регистра и меняющая ' ' на '_' 

def rename_snake_style(col_old):
    col_new = col_old.lower()
    col_new = re.sub(r' ','_', col_new)
    return col_new
```

### Выбор признаков

В качестве **целевого признака определено значение последнего замера температуры стали**, при этом количество замеров может отличаться от партии к партии 

Основываясь на результатах анализа, определим следующий пул исходных признаков как пригодный для использования при моделировании в качестве обучающей выборки:

- Данные о подаче сыпучих материалов (объём) (df_data_arc['bulk_1':'bulk_15'])
- Данные о подаче проволочных материалов (объём) (df_data_wire['wire_1':'wire_9'])


Приведенных выше признаков явно не достаточно для формирования обучающей выборки, по этому сгенереруем дополнительные признаки, основываясь на исходных данных.


Исходя из физических соображений (закон сохранения энергии) перечислим основные факторы влияющие на целевой признак (температуру) объекта измерений, так температура (как параметр связанный с ) зависит от:

- характеристик и исходного состояния системы
- переданой системе энергии
- переданной энергии системой внешней среде 


в характеристики системы в нашем случае входят параметры, отражающие состояние объекта измерения, такие как:
- исходный объем материала (не содержится в переданных данных)
- хим.состав (не содержится в переданных данных)
- изменение хим. состава:
    - Данные о добавках (содержится в исходных данных)
    - Время подачи сыпучих материалов (от начала плавки партии)
    - Время подачи проволочных материалов (от начала плавки партии)
    - Изменение удельных теплопотерь врезультате добавки на единицу добавки
- температура:
        - Исходная температура
        - Предпоследнее значение температуры 
        - Время предпоследнего замера температуры
        - Время первого замера температуры партии
    

в характеристики переданной системе и диссипированной во внешнюю среду энергии можно включить следующие параметры:

- Средняя активная мощность
- Средняя реактивная мощность
- Мощность

- Удельные теплопотери (в единицу времени, по этапам)

#### Средняя реактивная мощность

Средняя реактивная мощность

=

Сумма реактивных мощностей (для партии) 

/

```python
df_6_2_7 = df_data_arc.groupby(by='key').mean().copy().drop('active_power',axis=1)

df_6_2_7['key'] = df_6_2_7.index
```

#### Мощность

**Методология рассчета:** 

Мощность 

= 

квадратный корнь из суммы квадратов активной и реактивной мощностей

```python
df_6_2_8 = df_data_arc.copy().groupby(by='key').sum()

df_6_2_8['power'] = df_6_2_8.eval(f'(sqrt(active_power**2 + reactive_power**2))')

df_6_2_8 = df_6_2_8.drop(['active_power','reactive_power'],axis=1)

df_6_2_8['key'] = df_6_2_8.index
```

#### Коэфициент мощности

**Методология рассчета:** 

Коэфициент мощности

= 

Активная мощность
/
Мощность 

```python
df_6_2_9 = df_data_arc.groupby(by='key').sum().copy()

df_6_2_9['power_coef'] = df_6_2_9.eval(f'active_power/(sqrt(active_power**2 + reactive_power**2))')

df_6_2_9 = df_6_2_9.drop(['active_power','reactive_power'],axis=1)

df_6_2_9['key'] = df_6_2_9.index
```

#### Длительность обработки партии (длительность нагрева)

**Методология рассчета:** 

Длительность обработки партии (длительность нагрева) = сумма разностей времен начала нагрева и окончания нагрева (для партии)

```python
df_6_2_10 = df_data_arc[['key','start_arc_heating','end_arc_heating']].copy()

df_6_2_10.loc[:,'start_arc_heating'] = df_6_2_10.loc[:,'start_arc_heating'].dt.minute
df_6_2_10.loc[:,'end_arc_heating'] = df_6_2_10.loc[:,'end_arc_heating'].dt.minute


df_6_2_10['heat_duration'] = df_6_2_10.eval('(end_arc_heating - start_arc_heating)')

df_6_2_10 = df_6_2_10[['key','heat_duration']].groupby('key').sum()

df_6_2_10['key'] = df_6_2_10.index
```

- хим.состав (не содержится в переданных данных)
- изменение хим. состава:
    - Данные о добавках (содержится в исходных данных)
    - Время подачи сыпучих материалов (от начала плавки партии)
    - Время подачи проволочных материалов (от начала плавки партии)
    - Изменение удельных теплопотерь врезультате добавки на единицу добавки
- температура:
        - Исходная температура
        - Время первого замера температуры партии
    

в характеристики переданной системе и диссипированной во внешнюю среду энергии можно включить следующие параметры:

- Средняя активная мощность
- Средняя реактивная мощность
- Мощность
- Коэфициент мощности (активная мощность/мощность)
- Длительность обработки партии (длительность нагрева)
- Общее время плавки партии


судя по результатам анализа исходное сырье может быть неоднородным и его хим. состав может отличаться от партии к партии, используя для обучения только средние значения модель не будет "знакома" с партиями, имеющими подобное отличие, что негативно скажется на точности предсказания, по этому наряду со средними значениями параметров использованы и значения "в моменте".

Сформированы следующие признаки для обучения модели:
   - данные о добавках (объемы и типы сыпучих и проволочных материалов)
   - время подачи сыпучх материалов
   - время подачи проволочных материалов
   - исходная температура
   - общее время плавки
   - средняя активная мощность
   - средняя реактивная мощность
   - мощность
   - коэфициент мощности
   - длительность обработки партии
   
**При этом в случае не достижения моделью требуемого заказчиком уровня качества, рассматривался вопрос расширения пула признаков за счет добавления признака, отражающего изменение скорости остывания продукта после добавления той или иной добавки, однако при этом для его рассчета потребуется использование промежуточных температур, влияние этих данных на работу модели требует дополнительного исследования (так для снижения негативного влияния возможно провести усреднение данного параметра между всеми партиями). Гипотеза о возможности применения данного параметра основана на том, что добавление в продукт примесей приводит к изменению хим.состава и потенциально к изменению проводимости материала (коэфициента теплопередачи), что будет сказываться на скорости остывания**

рассмотрение корреляции подготовленных признаков показало отсутствие связи между отдельными параметрами (данные о подачи проволочных материалов из податчика номер 5) и целевым признаком.

### Моделирование

В рамках подготовки признаков произведено разделение общего пула признаков на обучающую и тестовую выборки в пропорции 75/25.

Для обучения выбраны следующие модели:
    - GradientBoostingRegressor
    - CatBoostRegressor
    - LinearRegression
    
Для каждой из моделей (за исключением модели линейной регрессии) проведен подбор оптимальных гиперпараметров.

Оценки работы моделей (на обучающей выборке) получены с применением кросс-валидации.

После произведен выбор модели, показавшей лучшие результаты (CatBoostRegressor) и произведено сравнение с результатами DummyRegressor (strategy="mean").

### Результаты

Точность прогноза модели, показавшей лучшие результаты на обучающей выборке (CatBoostRegressor) соответствет значению метрики mean absolute error (MAE) = 5.96 , что соответствет минимальному уровню качества ожидаемого заказчиком.

### Обсуждение полученных результатов

Непосредственное применение модели, обученной в ходе настоящего исследования позволяет прогнозировать температуру продукта при различном наборе исходных параметров, что позволит проводить их оптимизацию перед непосредственной обработкой продукта.

Так выбор парметров минимизирующих количество иттераций нагрева/охлаждения позволит снизить затраты на электронергию заказчика.

### Альтернативный взгляд на проблему

Помимо рассмотренного в рамках настоящего исследования способа решения конечной цели заказчика, возможно сформулировать и иные подходы, такие как:

- в случае наличия на производстве заказчика нескольких ковшей и возможности смешивания продукта в машине непрервной разливки можно предложить заказчику рассмотреть возможность смешивать партии из разных ковшей, не добиваясь эталонных значений хим. состава в каждом ковше но достигая его при их смешивании (данная мера позволит снизить количество иттераций нагрева/охлаждения и так же позволит снизить прямые затраты заказчика, однако при этом потребуется оценка технической возможности и потенциальной стоимость переоборудования производства)
- при решении задачи моделирования изменения теплопотерь продукта в зависимости от хим. состава возможно рассмотреть такой процесс обработки, при котором нагрев проиходит только один раз, а примеси (в случае необходимости их добавления при определенной температуре продукта) добавляются в расчитаных точках достижения необходимых температур 
- осуществлять не полный слив продукта из ковша, что может снизить потери энергии на первичный нагрев новой партии продукта в остывшем ковше
- подогрев добавляемых материалов за счет отдаваемого продуктом тепла позволит снизить энергию необходимую для донагрева продукта
